```python
import numpy as np
import pandas as pd
path ="/final-employee-ds.csv"
df = pd.read_csv(path)
```

Answer of Q1

```python
training_hours =df["TrainingHours"].to_numpy(dtype=float)
mean_val = np.mean(training_hours)
std_val = np.std(training_hours)
print(mean_val)
print(std_val)
```

ans of Q2

```python
df["Salary"] = df.groupby("Department")["Salary"].transform(
    lambda x: x.fillna(x.median()))

print(df.loc[
    (df.ExperienceYears > 12) & (df.ProjectCount >5),
    "Salary"].sum())
```

Ans of Q3

```python
result = df [
    ((df["Department"] == "IT")| (df["Department"] =="Finance")) &
    (df["Age"].between(30,45)) & (df["PerformanceScore"] > 88)
].sort_values("Salary", ascending = False)

print(result)
```

Answer of Q4

```python
df["SalaryPerHours"] = df["Salary"] / (df["WorkHoursPerWeek"]*4)

top5 = df.nlargest(5,"SalaryPerHours")[["Name","SalaryPerHours"]]

print(top5)
```
