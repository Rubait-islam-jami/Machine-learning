## Section 10 — Pairplot and Correlation Analysis

Explore relationships between features and correlations.


```python
# TODO:
# 1. Create a pairplot with selected numeric features plus the target
# 2. Generate a correlation matrix for numeric features and target
# 3. Visualize the matrix with a heatmap
# 4. Sort correlations with the target column

corr_matrix = df[numeric_cols + [target_col]].corr()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap for Numerical Features")
plt.tight_layout()
plt.show()
```
