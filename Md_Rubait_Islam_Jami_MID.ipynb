Sample Data

```python
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
"id": ids,
"age": ages,
"salary": salaries,
"dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
```

Q1

```python
data= np.genfromtxt('employees.csv', delimiter=",", dtype=None, names=True ,encoding =None)

ages_arr= data['age']
salaries_arr= data['salary']

print("Ages Array:",ages_arr)
print("dtype:",ages_arr.dtype)
print("ndim:",ages_arr.ndim)
print("shape:",ages_arr.shape)
print("size:",ages_arr.size)

print()

print("Salaries Array:",salaries_arr)
print("dtype:",salaries_arr.dtype)
print("ndim:",salaries_arr.ndim)
print("shape:",salaries_arr.shape)
print("size:",salaries_arr.size)


```

Q2

```python
print("Highest salary:", np.max(salaries_arr))
print("Lowest salary:", np.min(salaries_arr))
print("Average salary:", np.mean(salaries_arr))
print("Average age",np.mean(ages_arr))
```

Q3

```python
ages_greater_30 = ages_arr[ages_arr >30]

print("Ages greater than 30:",ages_greater_30)
print("Number of employees older than 30:",ages_greater_30.size)
```

Q4

```python
updated_ages = ages_arr + 5
print("Updated ages posivtly 5:",updated_ages)
```

Q5

```python
total_salary = np.sum(salaries_arr)
salary_range = np.max(salaries_arr)-np.min(salaries_arr)

print("Total salary expense:",total_salary)
print("Salary difference",salary_range)
```
